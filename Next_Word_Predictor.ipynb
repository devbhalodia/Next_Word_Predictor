```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs]) #applies certain modifications to the words, and assigns them integer indices.
```

```python
len(tokenizer.word_index) #gives total number of unique words
```

```python
'''
we have the dataset. now, we have to perform operations in each sentence. we have to then convert each sentence into i/p o/p format.
so, firstly, we split the sentence on the basis of new line.
to each sentence, we place the indices of words in place of words.
[sentence] gives us [[1,2,3]] which is a 2d list. we don't need a 2d list but just a 1d list. hence we did senetence[0].
so, each sentence got stored in tokenized_sentence one by one. so, say a particular sentence [1,2,3] is inside the tokenized_sentence.
now, in the next loop, this sentence converted into different input sequences and all these sequences got appended into the input_seq variable.
as in:-
[1,2] 1 being input and 2 being output
[1,2,3] 1,2 being input and 3 being output
'''
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences]) #finding max len for padding, as all the inputs are of different lengths and we dont feed different lengthed inputs to fully connected layers.
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre') #pre padding done
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
'''
if we see carefully into the input sequences, in each of the sequence has last char as the output and all the other char being input.
so, all the char except last, from all the rows, stored inside X.
'''
```

```python
y = padded_input_sequences[:,-1] #only the last char from all the rows stored in Y
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=283)
'''
total unique words are 282 but num_classes are 283 as it starts from 1 and not 0.
'''
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
'''
in Embedding, 283 words being unique, taken, and the output_dim being 100, all the words converted to 100 dim vectors.
'''
model.add(Embedding(283, 100, input_length=56))
'''
before, we had 3d vectors in the lstm variables like Ct, Ht, and so on. but for this problem, we have kept 150d vectors.
so, as we know, dim of Ct, ht, Ft, It, Ot all are same, being 150.
in the LTSM architecture, as each sentence has 56 words, there will be 56 timesteps.
so, only at the last time step, we will get our final output being H56 which will be of dim 150, that is, an array of 150 numbers.
'''
model.add(LSTM(150,return_sequences=True))
model.add(LSTM(150))
'''
this Ht is fed to the final output layer and we finally get the output out of 283 different classes.
'''
model.add(Dense(283, activation='softmax')) #output layer
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
import time
import numpy as np
text = "what is the fee"

for i in range(10): #10 iterations because we want to predict next 10 words based on the input text.
  #all the neccessary operations done on the text.
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  pos = np.argmax(model.predict(padded_token_text))
  '''
  pos being the position of the word in the dictionary. so, np.argmax converts the prediction into the number between 0 and 282.
  now, say we get pos = 49. so, the word at the index 49 taken and that is our next word.
  '''

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)

'''
what is tokenizer.word_index.items()???
text = "I love deep learning and machine learning"
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
print(tokenizer.word_index)
output:-
{'learning': 1, 'i': 2, 'love': 3, 'deep': 4, 'and': 5, 'machine': 6}
hence, say the new text was "i love" as the input:-
we apply all the tokenization operations and predict. say after predicition, pos = 3.
so, from the tokenizer.word_index.items(), we find the correspoding word at the index = pos = 3, which is "love" and this will become the next word.
now the new text is "i love machine" and we repeat the process for the next word till the first loop ends.
'''
```
